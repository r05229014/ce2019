```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('../data/pm25_2015.csv')
df.shape
```

```python
# 全台空氣指標 用空氣品質不好的測站數目當作評定標準

t = df['date']
pm25_50 = df.iloc[:, 2:]>50  # pm2.5 > 50 （當作空氣不好的指標）
Station_bad_50 = pm25_50.sum(axis=1) # pm25 > 50 的測站數量 
percent_station_bad = Station_bad_50*100/76  # 換成比例  （空氣不好的測站數量/全部測站）

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)

plt.figure(figsize=(12,8))
plt.plot(percent_station_bad)
plt.xlim(0,8760)
plt.ylim(0,100)
plt.ylabel('Bad air Stattion [%]')
plt.xlabel('hour')
```

```python
# 定義事件 連續6小時超過20%的測站為空氣不好的 當作一個"事件"
greater20 = percent_station_bad > 20 # >20%空氣不好
count = 0
memeory = []
for i in range(len(t)):
    if greater20[i] ==1:
        count += 1
        memeory.append(count)
    else:
        memeory.append(count)
        count = 0
target = np.zeros(len(memeory)).astype(int)
for i in range(len(t)):
    if memeory[i] == 6:
        target[i-5:i+1] = 1 
    elif memeory[i] > 6:
        target[i] = 1
c = Counter(memeory)
```

```python
event_count = target.sum()
pl = target*percent_station_bad
target2 = [1-val for val in target]
pl2 = target2*percent_station_bad

plt.figure(figsize=(12,8))
plt.plot(pl,'r', label = 'event')
plt.plot(pl2, 'b', label = 'not event')
plt.text(4000, 60,'event hours : %s hours' %event_count )
plt.text(4000, 75,'Count events : %s' %c[6] )
plt.legend()
plt.xlim(0,8760)
plt.ylim(0,100)
plt.ylabel('Bad air Stattion [%]')
plt.xlabel('hour')
```
